```python
class Human:
  def __init__(self, name, age):
    self.name = name
    self.age = age
  def check_adult(self):
    if self.age >= 20:
      print(f"{self.name}は大人です")
    else:
      print(f"{self.name}は大人ではありません")

person1 = Human("太郎", 15)
person2 = Human("花子", 25)
person3 = Human("健太", 19)

people_list = [person1, person2, person3]

for person in people_list:
  person.check_adult()
```
